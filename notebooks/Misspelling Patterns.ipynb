```python
import json
import pandas as pd

def load_jsonl(fname):
    fin = open(fname, encoding="utf-8")
    data = []
    for line in fin:
        d = json.loads(line.strip())
        data.append(d)

    return data

def save_jsonl(data, filename):
    with open(filename, "w", encoding="utf-8") as fo:
        for idx, d in enumerate(data):
            fo.write(json.dumps(d, ensure_ascii=False))
            fo.write("\n")
```

```python
sents = load_jsonl(f"../Datasets/WisesightSentiment/tokenized_test-misp.jsonl")
```

```python
from collections import defaultdict

cnt = defaultdict(int)
for s in sents:
    for seg in s["segments"]:
        for t1, t2 in zip(seg[0], seg[1]):
            cnt["token"] += 1
            if t1!=t2:
                cnt["misp"] += 1
            
print("%Misspelling", cnt["misp"]*100/cnt["token"], cnt["misp"])
```
